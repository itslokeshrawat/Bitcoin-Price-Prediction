```python
#Import the libraries
import numpy as np 
import pandas as pd
```

```python
#Load the dataset
from google.colab import files 
uploaded = files.upload() 
```

```python
#Store the data into the variable df
df = pd.read_csv('BTC-USD.csv')
df.head(5)
```

```python
#Remove the useless columns
df.drop(['Date','High','Low','Close','Adj Close','Volume'], 1, inplace=True)
```

```python
#Show the first 8 rows of the new data set
df.head(8)
```

```python
#A variable for predicting 30 days out into the future
prediction_days = 30

#Create another column shifted 'n' units up
df['Prediction'] = df[['Open']].shift(-prediction_days)
```

```python
#Show the first 8 rows of the new data set
df.head(8)
```

```python
#Show the last 8 rows of the new data set
df.tail(8)
```

```python
#CREATE THE INDEPENDENT DATA SET (X)

# Convert the dataframe to a numpy array and drop the prediction column
X = np.array(df.drop(['Prediction'],1))

#Remove the last 'n' rows where 'n' is the prediction_days
X= X[:len(df)-prediction_days]
print(X)
```

```python
#CREATE THE DEPENDENT DATA SET (y) 

# Convert the dataframe to a numpy array
y = np.array(df['Prediction'])  

# Get all of the y values except the last 'n' rows 
y = y[:-prediction_days] 
print(y)
```

```python
# Split the data into 80% training and 20% testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Set prediction_days_array equal to the last 30 rows of the original data set from the price column
prediction_days_array = np.array(df.drop(['Prediction'],1))[-prediction_days:]
print(prediction_days_array)
```

```python
from sklearn.svm import SVR
# Create and train the Support Vector Machine 
#Create the model
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
#Train the model
svr_rbf.fit(x_train, y_train) 
```

```python
# Testing Model accuracy on testing data
svr_rbf_score = svr_rbf.score(x_test, y_test)
print("svr_rbf accuracy: ", svr_rbf_score)
```

```python
# Print the predicted value for test dataset
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)
print()
```

```python
#Print the actual values
print(y_test)
```

```python
# Print the model predictions for the next '30' days
svm_prediction = svr_rbf.predict(prediction_days_array)
print(svm_prediction)
```
